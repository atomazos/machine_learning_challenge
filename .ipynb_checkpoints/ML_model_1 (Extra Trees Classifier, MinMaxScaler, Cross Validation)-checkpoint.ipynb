```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import ExtraTreesClassifier
from sklearn import metrics
```

### Read CSV and Clean Data

```python
df = pd.read_csv("exoplanet_data.csv")
# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')
# # Drop the null rows
df = df.dropna()
df.head()
df.koi_disposition.value_counts()
```

```python
y = df["koi_disposition"].apply(lambda x:0 if x == "FALSE POSITIVE" else 1)
```

```python
X = df.drop("koi_disposition", axis=1)
y.value_counts()
```

### Feature Selection (using Extra Trees Classifier and Random Forest Classifier)

```python
model = ExtraTreesClassifier(n_estimators=100, random_state =0) 
model.fit(X,y) 
```

```python
model.feature_importances_ 
```

```python
X.columns.to_list()
```

```python
feature_names = np.array(X.columns.to_list())
sorted_index = model.feature_importances_.argsort()
feature_importances = model.feature_importances_
```

```python
y_ticks = np.arange(0, len(feature_names))
fig, ax = plt.subplots(figsize=(10, 10))
ax.barh(y_ticks,feature_importances[sorted_index])
ax.set_yticklabels(feature_names[sorted_index])
ax.set_yticks(y_ticks)
ax.set_title("Extra Trees Classifier Feature Importances (MDI)")
fig.tight_layout()
plt.show()
```

```python
X = df[['koi_fpflag_ss', 'koi_fpflag_co', 'koi_fpflag_nt', 'koi_fpflag_ec']].copy()
```

### Create a Train - Test Split (with the selected features)

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
X_train     
```

### MinMaxScaler

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
# Fit on training set only.
scaler.fit(X_train)
# Apply transform to both the training set and the test set.
X_train = scaler.transform(X_train)
X_test= scaler.transform(X_test)
```

### Train the Model

```python
model.fit(X_train, y_train)
predictions = model.predict(X_test)
```

```python
print(f"Training Data Score for Extra Trees: {model.score(X_train, y_train)}")
print(f"Testing Data Score for Extra Trees: {model.score(X_test, y_test)}")
```

### Cross-validation

```python
clf = ExtraTreesClassifier(n_estimators=100, max_depth= None, min_samples_split=2,bootstrap=False, random_state=0)
scores = cross_val_score(clf, X_test, y_test, cv=5)
print(scores)
print(scores.mean())
```

```python
print("Average Cross Validation Score of ExtraTreesClassifier is 99%")
```

### Classification Report

```python
from sklearn.metrics import classification_report
target_names = ["Candidate", "Not"]
predictions = model.predict(X_test)
print(classification_report(y_test, predictions,
                            target_names=target_names))
```
